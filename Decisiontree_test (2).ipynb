```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data={
    'Age':[22,25,30,35,40,50],
    'Income':['Low','Medium','High','Low','High','Medium'],
    'Buys_Product':['No','Yes','Yes','No','Yes','No']
}
```

```python
df=pd.DataFrame(data)
```

```python
label_encoder_age=LabelEncoder()
label_encoder_income=LabelEncoder()
label_encoder_target=LabelEncoder()
```

```python
df['Age']=label_encoder_age.fit_transform(df['Age'])
df['Income']=label_encoder_income.fit_transform(df['Income'])
df['Buys_Product']=label_encoder_target.fit_transform(df['Buys_Product'])
```

```python
x=df[['Age','Income']]
y=df[['Buys_Product']]
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=42)
clf.fit(x,y)
```

```python
import matplotlib.pyplot as plt
plt.figsize=(10,6)
tree.plot_tree(clf,feature_names=['Age','Income'],class_names=['No','Yes'],filled=True)
plt.show()
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
data = {
    'Pclass': [1, 3, 3, 2, 1],
    'Sex': ['male', 'female', 'female', 'female', 'male'],
    'Age': [22, 38, 26, 35, 35],
    'SibSp': [1, 1, 0, 1, 0],
    'Parch': [0, 0, 0, 0, 0],
    'Fare': [71.2833, 7.925, 7.925, 53.1, 8.05],
    'Embarked': ['C', 'C', 'S', 'S', 'S'],
    'Survived': [0, 1, 1, 1, 0]
}
titanic_df = pd.DataFrame(data)
```

```python
le = LabelEncoder()
titanic_df['Sex'] = le.fit_transform(titanic_df['Sex'])
titanic_df['Embarked'] = le.fit_transform(titanic_df['Embarked'])
```

```python
X = titanic_df.drop('Survived', axis=1)
y = titanic_df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=X.columns, class_names=['Did not Survive', 'Survived'], filled=True)
plt.show()
tree_rules = export_text(clf, feature_names=X.columns.tolist())
print(tree_rules)
```
