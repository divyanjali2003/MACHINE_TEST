```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data={
    'Age':[22,25,30,35,40,50],
    'Income':['Low','Medium','High','Low','High','Medium'],
    'Buys_Product':['No','Yes','Yes','No','Yes','No']
}
```

```python
df=pd.DataFrame(data)
```

```python
label_encoder_age=LabelEncoder()
label_encoder_income=LabelEncoder()
label_encoder_target=LabelEncoder()
```

```python
df['Age']=label_encoder_age.fit_transform(df['Age'])
df['Income']=label_encoder_income.fit_transform(df['Income'])
df['Buys_Product']=label_encoder_target.fit_transform(df['Buys_Product'])
```

```python
x=df[['Age','Income']]
y=df[['Buys_Product']]
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=42)
clf.fit(x,y)
```

```python
import matplotlib.pyplot as plt
plt.figsize=(10,6)
tree.plot_tree(clf,feature_names=['Age','Income'],class_names=['No','Yes'],filled=True)
plt.show()
```

```python
company={
    'Days':[1,2,3,4,5,6],
    'Temperature':['Low','Medium','High','Low','High','Medium'],
    'Rain':['No','Yes','No','Yes','No','Yes']
}
```

```python
df1=pd.DataFrame(company)
```

```python
label_encoder_days=LabelEncoder()
label_encoder_temperature=LabelEncoder()
label_encoder_rain=LabelEncoder()
```

```python
df1['Days']=label_encoder_days.fit_transform(df1['Days'])
df1['Temperature']=label_encoder_temperature.fit_transform(df1['Temperature'])
df1['Rain']=label_encoder_rain.fit_transform(df1['Rain'])
```

```python
x=df1[['Days','Temperature']]
y=df1[['Rain']]
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=42)
clf.fit(x,y)
```

```python
import matplotlib.pyplot as plt
plt.figsize=(10,6)
tree.plot_tree(clf,feature_names=['Days','Temperature|'],class_names=['No','Yes'],filled=True)
plt.show()
```
